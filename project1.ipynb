```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#Creating purchase_frequency_days

frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Quarterly': 90, 
    'Annually': 365,
    'Fortnightly': 14, 
    'Every 3 months': 90 
}

df['purchase_frequency_days'] = df['Frequency of Purchases'].map(frequency_mapping)

df[['Frequency of Purchases', 'purchase_frequency_days']].head(10)
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']

df['age_group'] = pd.qcut(df['Age'], q=4, labels=labels)

df[['Age', 'age_group']].head(10)
```
